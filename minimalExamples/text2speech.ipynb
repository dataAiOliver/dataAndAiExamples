```python
import requests, os
from dotenv import load_dotenv
from IPython.display import Audio

load_dotenv()

HUGGINGFACE_API_TOKEN = os.getenv("HUGGINGFACE_API_TOKEN")
print(f"HUGGINGFACE_API_TOKEN: {HUGGINGFACE_API_TOKEN}")

def query(payload, API_URL = "https://api-inference.huggingface.co/models/espnet/kan-bayashi_ljspeech_vits"):
	headers = {"Authorization": f'Bearer {os.getenv("HUGGINGFACE_API_TOKEN")}'}
	response = requests.post(API_URL, headers=headers, json=payload)
	print(response)
	print(response.content)
	return response.content
	
audio = query({
	"inputs": "The answer to the universe is 42.",
})

with open("out/audio.flac", "wb") as f:
	f.write(audio)
	
Audio(audio, autoplay=True)
```
